```python
import pandas as pd

temp = pd.read_csv("/Users/anusreev/Desktop/python coding/temparature/tpweather.csv", encoding='latin1')

print(temp.head())
```

```python

days = temp['Day']
temperatures = temp['Temp']

plt.figure(figsize=(8, 5))
plt.plot(days, temperatures, marker='o', linestyle='-', color='Blue', linewidth=2)

plt.grid(True, linestyle='--')

plt.xlabel('Day')
plt.ylabel('Temperature (°C)')
plt.title('Temperature Trend Over the Last 5 Days')

plt.tight_layout()
plt.show()
```

```python

days = temp['Day']
temperatures = temp['Temp']

plt.figure(figsize=(8, 5))

plt.scatter(days, temperatures, color='blue', marker='o', label='Temperature')

plt.grid(True, linestyle='--')
plt.xlabel('Day')
plt.ylabel('Temperature (°C)')
plt.title('Temperature Trend Over the Last 5 Days')

plt.tight_layout()
plt.legend()
plt.show()
```
