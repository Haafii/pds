```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('StudentsPerformance.csv')

# Set the figure size
plt.figure(figsize=(12, 6))

sns.heatmap(df.isnull(), cbar=False, cmap='viridis', yticklabels=False)
plt.title('Heatmap of Missing Values')
plt.show()

```

```python
df.isnull().sum()
```

```python


# Fill numeric columns with mean
numeric_cols = df.select_dtypes(include=['number']).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())

# Fill non-numeric columns with 0
non_numeric_cols = df.select_dtypes(exclude=['number']).columns
df[non_numeric_cols] = df[non_numeric_cols].fillna(0)


print(df.head())
```

```python
df.isnull().sum()
```

```python
#Visualize outliers using scatterplot
# Math vs Reading
sns.scatterplot(data=df, x='math score', y='reading score')
plt.title('Math vs Reading Score')
plt.show()

# Math vs Writing
sns.scatterplot(data=df, x='math score', y='writing score')
plt.title('Math vs Writing Score')
plt.show()

# Reading vs Writing
sns.scatterplot(data=df, x='reading score', y='writing score')
plt.title('Reading vs Writing Score')
plt.show()
```

```python
#Histograms to Explore Score Distributions

# Set Seaborn style
sns.set(style="whitegrid")

# Create subplots for each score
plt.figure(figsize=(18, 5))

# Histogram for Math Score
plt.subplot(1, 3, 1)
sns.histplot(df['math score'], bins=20, kde=True, color='skyblue')
plt.title('Math Score Distribution')
plt.xlabel('Math Score')
```

```python
plt.figure(figsize=(18, 5))

plt.subplot(1, 3, 2)
sns.histplot(df['reading score'], bins=20, kde=True, color='skyblue')
plt.title('Reading Score Distribution')
plt.xlabel('Reading Score')
```

```python
plt.figure(figsize=(18, 5))

plt.subplot(1, 3, 3)
sns.histplot(df['writing score'], bins=20, kde=True, color='skyblue')
plt.title('Writing Score Distribution')
plt.xlabel('Writing Score')
```

```python
#Compare scores by gender or test prep coursewith boxplot or violinplot
plt.subplot(1, 3, 1)
sns.boxplot(x='gender', y='math score', data=df, palette='Set2')
plt.title('Math Score by Gender')
```

```python
plt.subplot(1, 3, 2)
sns.boxplot(x='gender', y='reading score', data=df, palette='Set2')
plt.title('Reading Score by Gender')
```

```python
plt.subplot(1, 3, 3)
sns.boxplot(x='gender', y='writing score', data=df, palette='Set2')
plt.title('Writing Score by Gender')
```

```python
#calculate correlation matrix
corr_matrix = df.corr(numeric_only=True)
print("Correlation Matrix:\n")
print(corr_matrix)
```

```python
#Visualize Corelation using HeatMap
sns.heatmap(corr_matrix, annot=True, linewidths=0.5, square=True)
plt.show()
```

```python
#Draw the bar graph using Matplotlib for average gender scores
#a. df = pd.read_csv("StudentsPerformance.csv")
#b. gender_scores = df.groupby("gender")[["math score", "reading score",
#"writing score"]].mean()
gender_scores = df.groupby("gender")[["math score", "reading score", "writing score"]].mean()
gender_scores.T.plot(kind='bar', figsize=(10, 6), rot=0)
plt.title("Average Scores by Gender")
plt.ylabel("Average Score")
plt.xlabel("Subjects")
plt.title("Average Scores by Gender")
plt.show()
```
